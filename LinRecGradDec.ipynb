## Step 2 Using analytical approach we solve this using calculus:

### Take the derivative of the cost function, this is a single param problem looking at theta so it will it in respect to theta

# taking the derivative using chain rule gives us 4(2(theta) - 3) 

## Step 3 find critical point 
# Using the derivative above we can find the critical point by setting it equal to 0 and solving

# 4(2(theta) - 3) = 0 gives the point give us 3/2 which

## Step 4 plugging back into the equation to find the y coordinate

```python
import numpy as np
import matplotlib.pyplot as plt


def J(theta):
    return (2 * theta - 3) ** 2

theta_values = np.linspace(-1, 3, 100)
J_values = J(theta_values)

# Plot J(θ) vs θ
plt.figure(figsize=(8, 5))
plt.plot(theta_values, J_values, label='J(θ) = (2θ - 3)²', color='blue')
plt.title('Plot of J(θ) vs θ')
plt.xlabel('θ')
plt.ylabel('J(θ)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

# Using gradient descent to solve this problem

# Step 1:  the cost function is given and like above we need to find the derivative which is 4(2(theta) - 3) which we already found

## Step 2: now that we have dJ/dtheta we need to use this in the new formula for gradient descent and use iterations to get closer to the value that we need 

## Step 3: The gradient descent formula for the parameter to optimize is theta_new = theta_current - alpha dJ/dtheta 

## Step 4: Choose a starting point: For this I will choose theta to be 0 and alpha to be .1 

## Step 5: run the iteraiton until covergence

```python
def gradient(theta):
    return 4 * (2 * theta - 3)

theta = 5
alpha = 0.1  
tolerance = 1e-5
max_iterations = 100

theta_values = [theta]
for i in range(max_iterations):
    grad = gradient(theta)
    theta = theta - alpha * grad  
    theta_values.append(theta)

    print(f"Iteration {i+1}: θ = {theta:.6f}, J(θ) = {J(theta):.6f}")

    if abs(grad) < tolerance:
        break

print("\nFinal Results:")
print(f"Optimal θ: {theta:.6f}")
print(f"Minimum J(θ): {J(theta):.6f}")
print(f"Total Iterations: {len(theta_values) - 1}")
```
